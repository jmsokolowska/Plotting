```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fortune500-full (1).csv')
#  odczyt danych z pliku csv
```

```python
sns.set_context('paper')
sns.relplot(data=df[(df['Company'] == 'General Motors') | (df['Company'] == 'Esmark') | (df['Company'] == 'Honeywell') ],
            x="Year",
            y="Revenue (in millions)",
            aspect=2.5, 
            kind='line',
            hue='Company')
plt.show()
# wyświetlenie wykresu liniowego dla firm General Motors, Esmark i Honeywell z wykorzystaniem biblioteki seaborn
```

```python
sns.set_context('paper')
sns.lmplot(data=df[(df['Company'] == 'General Motors') | (df['Company'] == 'Esmark') | (df['Company'] == 'Honeywell') ],
            x="Year",
            y="Revenue (in millions)",
            aspect=2.5, 
            hue='Company')
plt.show()
# modyfikacja powyższego wykresu - trend
```

```python
sns.set_context('poster')
sns.catplot(data=df[(df['Company'] == 'General Motors') | (df['Company'] == 'Esmark') | (df['Company'] == 'Honeywell') ],
            x="Year",
            y="Revenue (in millions)",
            aspect=10, 
            kind='bar',
            hue='Company')
plt.show()
# modyfikacja - histogram (dystrybucja)
```

```python
sns.set_context('poster')
sns.relplot(data=df[(df['Company'] == 'General Motors') | (df['Company'] == 'Esmark') | (df['Company'] == 'Honeywell') ],
        kind='scatter', 
        x='Year', 
        y='Revenue (in millions)', 
        aspect=7 )
plt.show()
# modyfikacja - wykres punktowy
```
